4)Use the dictionary 'n_users' to plot a pie chart as shown below.

image.png

```python
# run this cell before proceeding further
n_users = {'Windows':323,'Android':403,'iOS':165,'OS X':71,'Linux':9}
```

```python
#your code here
fig=plt.figure()
l=[]
colors=['#0abde3', '#C4E538', '#7f8c8d', '#bdc3c7','#f39c12']
d=n_users.values()
for i in n_users:
  l.append(i)
plt.pie(d,labels=l,autopct='%1.f%%',colors=colors)
plt.show()
```

1) Plot the heatmap of the correlation matrix of the iris data using seaborn.
Center the colorbar at 0 and display the data value in each cell.

```python
#your code here
import pandas as pd
iris=pd.read_csv("https://raw.githubusercontent.com/Spectrum-CETB/Data-Science-AI/main/Datasets/iris.csv")
iris.drop('Id',axis=1,inplace=True)
fig=plt.figure()
fig.set_size_inches(10,7)
fig=sns.heatmap(iris.corr(),annot=True,cmap="coolwarm",center=0)
```

2) Using the iris data draw a boxplot 
- horizontally for each numeric variable in the dataset.
- for SepalWidthCm	 grouped by Species.

```python
#your code here
import pandas as pd
iris=pd.read_csv("https://raw.githubusercontent.com/Spectrum-CETB/Data-Science-AI/main/Datasets/iris.csv")
iris.drop('Id',axis=1,inplace=True)
fig1=plt.figure(1)
fig1=sns.boxplot(data=iris)
fig2=plt.figure(2)
fig2=sns.boxplot(x='Species',y='SepalWidthCm',data=iris)
plt.show()
```

3)Plot multiple pairwise bivariate distributions of the above dataset by taking "Species" as the Variable  to map plot aspects to different colors.

```python
#your code here
import pandas as pd
iris=pd.read_csv("https://raw.githubusercontent.com/Spectrum-CETB/Data-Science-AI/main/Datasets/iris.csv")
iris.drop('Id',axis=1,inplace=True)
sns.pairplot(iris,hue='Species');
```

4)Replicate the below plot using seaborn library.

seaborn1.png

```python
#your code here
sns.set_style("darkgrid")
fig=sns.jointplot(x=iris["SepalLengthCm"], y=iris["SepalWidthCm"],data=iris,s=90)
```

5)Using seaborn replicate the plot below.

seaborn2.png

```python
#your code here.
fig=plt.gcf()
fig.set_size_inches(7,5)
sns.set_style("darkgrid")
colors=["#e84393","#27ae60","#2980b9"]
fig=sns.violinplot(x='Species',y='PetalLengthCm',data=iris,palette=colors)
```
